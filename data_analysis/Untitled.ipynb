```python
%matplotlib inline
import csv
import matplotlib.pyplot as plt
```

```python
sz = [10000, 10000000]
ac = [1000000, 100000]
skew = ["0.000000", "0.200000", "0.500000", "0.800000", "1.050000", "1.250000", "1.500000", "2.000000", "2.500000","3.000000","15.000000"]
```

```python
ll = []

d = []   
for s in skew:
    dt = []
    for i in range(10000+1):
        dt.append(0)
    d.append(dt)

c = 0
for s in skew:
    l = []
    name = str(10000) + "_" + str(1000000) + "_" + s + ".csv"
    with open(name, newline='') as csvfile:
        reader = csv.reader(csvfile)
        for row in reader:
            l.append(row[0])
            d[c][int(row[0])] = d[c][int(row[0])] + 1
        c += 1
    print(s, len(l))
    ll.append(l)
print( len(ll))        
    
    
```
